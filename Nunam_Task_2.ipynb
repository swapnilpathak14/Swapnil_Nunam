#### Resampling to 1 sample per minute

```python
!pip install xlrd
```

```python
df_new['Absolute Time'] = df_new['Absolute Time'].astype('datetime64')
df_new_1['Realtime'] = df_new_1['Realtime'].astype('datetime64')
df_new_2['Realtime'] = df_new_2['Realtime'].astype('datetime64')
```

##### Downsampled

```python
df_new_t = df_new.resample('60s', on='Absolute Time').first()
```

```python
df_new_t1 = df_new_1.resample('60s', on='Realtime').first()
```

```python
df_new_t2 = df_new_2.resample('60s', on='Realtime').first()
```
